```python
import pandas as pd


data = pd.read_csv('Quantitative_data.csv', low_memory=False)
```

```python
# data2 = pd.read_csv('dimension_reduction_datasets/UMAPminmaxScaler-simpleRegressor.csv', low_memory=False)
data2 = pd.read_csv('dimension_reduction_datasets/t-SNEminmaxScaler-simpleRegressor.csv', low_memory=False)
# data2 = pd.read_csv('dimension_reduction_datasets/UMAPminmaxScaler-simpleRegressor.csv', low_memory=False)
# data2 = pd.read_csv('dimension_reduction_datasets/UMAPminmaxScaler-simpleRegressor.csv', low_memory=False)
# data2 = pd.read_csv('dimension_reduction_datasets/UMAPminmaxScaler-simpleRegressor.csv', low_memory=False)
# data2 = pd.read_csv('dimension_reduction_datasets/UMAPminmaxScaler-simpleRegressor.csv', low_memory=False)
# data2 = pd.read_csv('dimension_reduction_datasets/UMAPminmaxScaler-simpleRegressor.csv', low_memory=False)
# data2 = pd.read_csv('dimension_reduction_datasets/UMAPminmaxScaler-simpleRegressor.csv', low_memory=False)
# data2 = pd.read_csv('dimension_reduction_datasets/UMAPminmaxScaler-simpleRegressor.csv', low_memory=False)
```

```python
# df = pd.concat([data2[['UMAP1', 'UMAP2']], data[['Group', 'Subgroup', 'Is Master Protein?', 'Name', 'Species', 'Taxonomic Domain']]], axis=1)
df = pd.concat([data2[['TSNE1', 'TSNE1']], data[['Group', 'Subgroup', 'Is Master Protein?', 'Name', 'Species', 'Taxonomic Domain']]], axis=1)
```

```python
mono_df = df[df['Subgroup'].isin(['Cyclooxygenases', 'Lipoxygenases'])]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a different color palette (Paired)
colors = sns.color_palette("Paired", n_colors=len(mono_df['Subgroup'].unique()))

# Create a scatter plot with larger markers
# sns.scatterplot(x='UMAP1', y='UMAP2', hue='Subgroup', style='Subgroup', data=mono_df, palette=colors, s=150)
sns.scatterplot(x='TSNE1', y='TSNE1', hue='Subgroup', style='Subgroup', data=mono_df, palette=colors, s=150)

# Label each point with its title
# for i, row in mono_df.iterrows():
#     plt.text(row['UMAP1'], row['UMAP2'], row['Subgroup'], ha='right', va='bottom')

# Add labels and title
plt.xlabel('UMAP1 Values')
plt.ylabel('UMAP2 Values')
plt.title('Scatter Plot with Labels')

# Show the plot
plt.show()
```
