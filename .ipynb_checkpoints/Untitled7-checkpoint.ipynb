```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, OPTICS, MeanShift, Birch, MiniBatchKMeans
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# Generate sample data
n_samples = 300
data = pd.read_csv("dimension_reduction_datasets/t-SNEmaxabsScaler-KNN.csv")

# Define a range of cluster numbers to test
k_values = range(2, 11)

# Initialize dictionaries to store results for different algorithms
results = {
    'K-Means': [],
    'Agglomerative': [],
    'DBSCAN': [],
    'GMM': [],
    'OPTICS': [],
    'MeanShift': [],
    'BIRCH': [],
    'MiniBatchKMeans': []
}

# Evaluate each clustering algorithm with different parameters
for k in k_values:
    # Define different parameter options for each algorithm
    algorithm_params = {
        'K-Means': [{'n_clusters': k, 'n_init': 10}, {'n_clusters': k, 'n_init': 20}],
        'Agglomerative': [{'n_clusters': k, 'linkage': 'ward'}, {'n_clusters': k, 'linkage': 'complete'}, {'n_clusters': k, 'linkage': 'average'}],
        'DBSCAN': [{'eps': 0.1, 'min_samples': 5}, {'eps': 0.3, 'min_samples': 5}, {'eps': 0.5, 'min_samples': 5}],
        'GMM': [{'n_components': k, 'covariance_type': 'full'}, {'n_components': k, 'covariance_type': 'tied'}],
        'OPTICS': [{'min_samples': 5, 'xi': 0.05, 'cluster_method': 'xi'}, {'min_samples': 10, 'xi': 0.1, 'cluster_method': 'xi'}],
        'MeanShift': [{'bandwidth': 0.5}, {'bandwidth': 1.0}, {'bandwidth': 1.5}],
        'BIRCH': [{'n_clusters': k}, {'n_clusters': k, 'branching_factor': 20}],
        'MiniBatchKMeans': [{'n_clusters': k, 'batch_size': 50}, {'n_clusters': k, 'batch_size': 100}]
    }
    for algorithm_name, param_options in algorithm_params.items():
        best_metric = float('inf')
        for params in param_options:
            algorithm = None
            if algorithm_name == 'K-Means':
                algorithm = KMeans(**params)
            elif algorithm_name == 'Agglomerative':
                algorithm = AgglomerativeClustering(**params)
            elif algorithm_name == 'DBSCAN':
                algorithm = DBSCAN(**params)
            elif algorithm_name == 'GMM':
                algorithm = GaussianMixture(**params)
            elif algorithm_name == 'OPTICS':
                algorithm = OPTICS(**params)
            elif algorithm_name == 'MeanShift':
                algorithm = MeanShift(**params)
            elif algorithm_name == 'BIRCH':
                algorithm = Birch(**params)
            elif algorithm_name == 'MiniBatchKMeans':
                algorithm = MiniBatchKMeans(**params)

            algorithm.fit(data)

            # Calculate and store the appropriate evaluation metric for each algorithm
            if algorithm_name in ['K-Means', 'MiniBatchKMeans']:
                metric = algorithm.inertia_
            elif algorithm_name == 'Agglomerative':
                from sklearn.metrics import silhouette_score
                labels = algorithm.labels_
                metric = silhouette_score(data, labels)
            elif algorithm_name == 'GMM':
                metric = -algorithm.lower_bound_  # Lower bound of the log-likelihood
            elif algorithm_name == 'DBSCAN':
                # You can use other evaluation metrics such as silhouette_score, calinski_harabasz_score, etc.
                from sklearn.metrics import silhouette_score
                labels = algorithm.labels_
                if len(set(labels)) > 1:
                    metric = silhouette_score(data, labels)
                else:
                    metric = -1  # Handle single-cluster cases

            if metric < best_metric:
                best_metric = metric

        results[algorithm_name].append(best_metric)

# Plot the results for each algorithm
plt.figure(figsize=(16, 8))
for algorithm_name, values in results.items():
    plt.plot(k_values, values, marker='o', label=algorithm_name)

plt.title("Elbow Method for Unsupervised Clustering Algorithms")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Evaluation Metric")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN, KMeans

# silhouette_score for evaluating the clusters, ranges from -1 to +1, desirable score is close to 1

from sklearn.metrics import silhouette_score
data = pd.read_csv("./dimension_reduction_datasets/FactorAnalysismaxabsScaler-KNN.csv")
sns.pairplot(data[["PCA1", "PCA2"]], height=2);

nn = NearestNeighbors(n_neighbors=20).fit(data[["PCA1", "PCA2"]])
distances, indices = nn.kneighbors(data[["PCA1", "PCA2"]])

distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.figure(figsize=(10,8))
plt.plot(distances)
```

```python
from collections import Counter
min_samples = range(10,21)
eps = np.arange(0.05,0.13, 0.01) # returns array of ranging from 0.05 to 0.13 with step of 0.01
output = []

for ms in min_samples:
    for ep in eps:
        labels = DBSCAN(min_samples=ms, eps = ep).fit(data[["PCA1", "PCA2"]]).labels_
        score = silhouette_score(data[["PCA1", "PCA2"]], labels)
        output.append((ms, ep, score))

min_samples, eps, score = sorted(output, key=lambda x:x[-1])[-1]
print(f"Best silhouette_score: {score}")
print(f"min_samples: {min_samples}")
print(f"eps: {eps}")

labels = DBSCAN(min_samples=min_samples, eps = 0.11).fit(data[["PCA1", "PCA2"]]).labels_
clusters = len(Counter(labels))
print(f"Number of clusters: {clusters}")
print(f"Number of outliers: {Counter(labels)[-1]}")
# print(f"Silhouette_score: {silhouette_score(data[["PCA1", "PCA2"]], labels)}")

plt.figure(figsize=(10,8))
sns.scatterplot(x=data["PCA1"], y=data["PCA2"], hue=labels, palette='viridis');
```
