```python
import pandas as pd
import ast

def convertColumn(x):
    content = ast.literal_eval(x) if isinstance(x, str) else x
    result = sum(content) / len(content) if isinstance(content, list) else content
    return result
data = pd.read_csv("Quantitative_data.csv")
# Split the string, convert to numeric values, and calculate the average
data['resolution_quantitative'] = data['rcsb_entry_info_resolution_combined'].apply(lambda x: convertColumn(x))

# Extract the header as a DataFrame with one row
numeric_column_names = data.select_dtypes(include='number').columns
df = data[numeric_column_names]
# remove _id columns
# Remove columns with "_id" in their headers
filtered_columns = [col for col in df.columns if '_id' not in col]

df[filtered_columns]
```

```python
colco = [col for col in df[filtered_columns].columns if '0' not in col]

Quantitative = df[colco]
Quantitative.to_csv("NewQuantitative.csv")
```
