```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd

df = pd.read_csv("./dimension_reduction_datasets/UMAPmaxabsScaler-KNN.csv")

data = list(zip(df["PCA1"], df["PCA2"]))

linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd

df = pd.read_csv("./dimension_reduction_datasets/t-SNEmaxabsScaler-KNN.csv")

data = list(zip(df["PCA1"], df["PCA2"]))

linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

df = pd.read_csv("./dimension_reduction_datasets/t-SNEmaxabsScaler-softImputerRegressor.csv")

data = list(zip(df["PCA1"], df["PCA2"]))

hierarchical_cluster = AgglomerativeClustering(n_clusters=10, affinity='euclidean', linkage='ward')
labels = hierarchical_cluster.fit_predict(data)

plt.scatter(df["PCA1"], df["PCA2"], c=labels, cmap='viridis')
plt.show() 
```
