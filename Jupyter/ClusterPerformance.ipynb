```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Generate some example data (you can replace this with your dataset)
data, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Initialize a dictionary to store the results for each algorithm
results = {}

# Define the range of clusters you want to try
num_clusters_range = range(2, 11)

# K-Means
kmeans_scores = {
    'Silhouette Score': [],
    'Calinski-Harabasz Score': [],
    'Davies-Bouldin Score': []
}

for num_clusters in num_clusters_range:
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans_labels = kmeans.fit_predict(data)
    
    silhouette = silhouette_score(data, kmeans_labels)
    calinski_harabasz = calinski_harabasz_score(data, kmeans_labels)
    davies_bouldin = davies_bouldin_score(data, kmeans_labels)
    
    kmeans_scores['Silhouette Score'].append(silhouette)
    kmeans_scores['Calinski-Harabasz Score'].append(calinski_harabasz)
    kmeans_scores['Davies-Bouldin Score'].append(davies_bouldin)

results['K-Means'] = kmeans_scores

# Agglomerative Hierarchical Clustering
agglo_scores = {
    'Silhouette Score': [],
    'Calinski-Harabasz Score': [],
    'Davies-Bouldin Score': []
}

for num_clusters in num_clusters_range:
    agglo = AgglomerativeClustering(n_clusters=num_clusters)
    agglo_labels = agglo.fit_predict(data)
    
    silhouette = silhouette_score(data, agglo_labels)
    calinski_harabasz = calinski_harabasz_score(data, agglo_labels)
    davies_bouldin = davies_bouldin_score(data, agglo_labels)
    
    agglo_scores['Silhouette Score'].append(silhouette)
    agglo_scores['Calinski-Harabasz Score'].append(calinski_harabasz)
    agglo_scores['Davies-Bouldin Score'].append(davies_bouldin)

results['Agglomerative'] = agglo_scores

# DBSCAN (Doesn't require specifying the number of clusters)
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data)
num_clusters_dbscan = len(np.unique(dbscan_labels))
results['DBSCAN'] = {'Number of Clusters': num_clusters_dbscan}

# Print the results
for algorithm, scores in results.items():
    print(f"{algorithm}:")
    if algorithm != 'DBSCAN':
        best_num_clusters = num_clusters_range[np.argmax(scores['Silhouette Score'])]
        print(f"Optimal Number of Clusters (Silhouette Score): {best_num_clusters}")
    
    for score_name, score_values in scores.items():
        if score_name != 'Number of Clusters':
            print(f"{score_name}: {score_values}")
    
    print("\n")

# Visualize the results (optional)
plt.figure(figsize=(12, 8))
for algorithm, scores in results.items():
    if algorithm != 'DBSCAN':
        plt.plot(num_clusters_range, scores['Silhouette Score'], label=f"{algorithm} - Silhouette Score")
    else:
        plt.axvline(x=num_clusters_dbscan, color='k', linestyle='--', label=f"{algorithm} - Estimated Clusters")

plt.title('Clustering Performance Metrics')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.xticks(num_clusters_range)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.cluster import OPTICS, MeanShift, AffinityPropagation
from sklearn.cluster import Birch, MiniBatchKMeans

# Generate some example data (you can replace this with your dataset)
data, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Initialize a dictionary to store the results for each algorithm
results = {}

# Define the range of clusters you want to try
num_clusters_range = range(2, 11)

datasets = {
    "simpleRegressor": [
        "standardScaler-simpleRegressor", 
        'minmaxScaler-simpleRegressor', 
        'robustScaler-simpleRegressor', 
        'maxabsScaler-simpleRegressor', 
        'unitVectorScaler-simpleRegressor', 
        'mquantileTransformScaler-simpleRegressor'
    ],
    "KNN": [
        "standardScaler-KNN", 
        'minmaxScaler-KNN', 
        'robustScaler-KNN', 
        'maxabsScaler-KNN', 
        'unitVectorScaler-KNN', 
        'mquantileTransformScaler-KNN'
    ],
    "softImputerRegressor": [
        "standardScaler-softImputerRegressor", 
        'minmaxScaler-softImputerRegressor', 
        'robustScaler-softImputerRegressor', 
        'maxabsScaler-softImputerRegressor', 
        'unitVectorScaler-softImputerRegressor', 
        'mquantileTransformScaler-softImputerRegressor'
    ],
    "iterativeImputerRegressor": [
        "standardScaler-iterativeImputerRegressor", 
        'minmaxScaler-iterativeImputerRegressor', 
        'robustScaler-iterativeImputerRegressor', 
        'maxabsScaler-iterativeImputerRegressor', 
        'unitVectorScaler-iterativeImputerRegressor', 
        'mquantileTransformScaler-iterativeImputerRegressor'
    ]
}


# Initialize dimensionality reduction techniques
dim_reduction_techniques = [
    'PCA', 'TruncatedSVD', 'FastICA', 'MDS', 't-SNE', 'UMAP', 'FactorAnalysis', 'Isomap'
]
        
# Define clustering algorithms
clustering_algorithms = {
    'K-Means': KMeans(n_init = 10),
    'Agglomerative': AgglomerativeClustering(),
    'DBSCAN': DBSCAN(),
    'GMM': GaussianMixture(),
    'OPTICS': OPTICS(),
    'MeanShift': MeanShift(),
    'BIRCH': Birch(),
    'MiniBatchKMeans': MiniBatchKMeans()
}
for i, group_data in enumerate(datasets):
    for dataset in datasets.get(group_data):
        create_dir("elbow_charts/" + group_data + "/")
        for technique_name in dim_reduction_techniques:
            check_file = "dimension_reduction_datasets/" + dataset+"_"+technique_name+"/"+ technique_name+dataset+".csv"
            if os.path.exists(check_file):
                data = pd.read_csv(check_file)

            check_chart_exist = "elbow_charts/"+ group_data + "/" +dataset+"_"+technique_name+"-high_quality_image.png"
            if(os.path.exists(check_chart_exist)):
                print("moving to the next")
                continue
            for algorithm_name, algorithm in clustering_algorithms.items():
                scores = {
                    'Silhouette Score': [],
                    'Calinski-Harabasz Score': [],
                    'Davies-Bouldin Score': []
                }
                
                for num_clusters in num_clusters_range:
                    if algorithm_name not in ['DBSCAN', 'OPTICS', 'MeanShift']:
                        algorithm.n_clusters = num_clusters
                    elif algorithm_name in ['DBSCAN', 'MeanShift']:
                        continue  # No need to set the number of clusters for these algorithms
                    
                    if algorithm_name != 'GMM':
                        labels = algorithm.fit_predict(data)
                    else:
                        algorithm.n_components = num_clusters
                        labels = algorithm.fit(data).predict(data)
                    
                    silhouette = silhouette_score(data, labels)
                    calinski_harabasz = calinski_harabasz_score(data, labels)
                    davies_bouldin = davies_bouldin_score(data, labels)
                    
                    scores['Silhouette Score'].append(silhouette)
                    scores['Calinski-Harabasz Score'].append(calinski_harabasz)
                    scores['Davies-Bouldin Score'].append(davies_bouldin)
                
                results[algorithm_name] = scores
            
            # Print the results
            for algorithm, scores in results.items():
                print(f"{algorithm}:")
                
                for score_name, score_values in scores.items():
                    print(f"{score_name}: {score_values}")
                
                print("\n")
            
            # Visualize the results (optional)
            plt.figure(figsize=(12, 8))
            for algorithm, scores in results.items():
                if algorithm not in ['DBSCAN', 'MeanShift', 'AffinityPropagation']:
                    plt.plot(num_clusters_range, scores['Silhouette Score'], label=f"{algorithm} - Silhouette Score")
            
            plt.title('Clustering Performance Metrics')
            plt.xlabel('Number of Clusters')
            plt.ylabel('Score')
            plt.xticks(num_clusters_range)
            plt.legend()
            plt.grid(True)
            # Specify a high dpi (e.g., 300) for high-quality image
            create_dir("elbow_charts/")
            plt.savefig("elbow_charts/"+ group_data + "/" +dataset+"_"+technique_name+'-high_quality_image.png', dpi=300, bbox_inches='tight')  # Save the figure
            plt.show()
```
