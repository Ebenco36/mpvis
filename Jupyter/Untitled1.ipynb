```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Define a list of clustering algorithms
clustering_algorithms = [
    ("K-Means", KMeans(n_clusters=4, random_state=42)),
    ("DBSCAN", DBSCAN(eps=0.5, min_samples=5)),
    ("Agglomerative", AgglomerativeClustering(n_clusters=4))
]

# Evaluate each clustering algorithm
for name, algorithm in clustering_algorithms:
    algorithm.fit(X)
    labels = algorithm.labels_

    # Silhouette Score
    silhouette_avg = silhouette_score(X, labels)
    print(f"{name} - Silhouette Score: {silhouette_avg:.2f}")

    # Davies-Bouldin Index
    davies_bouldin = davies_bouldin_score(X, labels)
    print(f"{name} - Davies-Bouldin Index: {davies_bouldin:.2f}")

    # Calinski-Harabasz Index
    calinski_harabasz = calinski_harabasz_score(X, labels)
    print(f"{name} - Calinski-Harabasz Index: {calinski_harabasz:.2f}")

    # Plot the clusters
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
    plt.title(f"{name} Clustering")
    plt.show()
```
